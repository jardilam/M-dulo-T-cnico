```python
# Descarga de datos de las acciones desde yahoo

ITServDf = yf.download("acn ibm fi", start="2002-01-01")
```

```python
# Convertir los datos en un DataFrame:

ITServDf = pd.DataFrame(ITServDf)
```

```python
# Guardar DataFrame en archivo .csv

ITServDf.to_csv("DataFrame ITServ", index=False)
```

```python
# Agregar columna Date

ITServDf["Date"] = pd.date_range(start="2002-01-01", periods=len(ITServDf), freq="D")

ITServDf
```

```python
# Agregar columna daily return

ITServDf["Daily Return ACN"] = ITServDf["Adj Close"]["ACN"].pct_change()

ITServDf

ITServDf["Daily Return FI"] = ITServDf["Adj Close"]["FI"].pct_change()

ITServDf

ITServDf["Daily Return IBM"] = ITServDf["Adj Close"]["IBM"].pct_change()

ITServDf
```

```python
# Verificar nulos

ITServDf.isnull().sum()
```

```python
# Calcula algunas estadísticas descriptivas básicas para el precio de cierre
print('Media del precio de cierre: ', ITServDf['Close'].mean())
print('Mediana del precio de cierre: ', ITServDf['Close'].median())
print('Desviación estándar del precio de cierre: ', ITServDf['Close'].std())
print('Mínimo del precio de cierre: ', ITServDf['Close'].min())
print('Máximo del precio de cierre: ', ITServDf['Close'].max())
```

```python
# Eliminar la primera fila porque tiene un valor nulo

ITServDf = ITServDf.dropna().reset_index(drop=True)
ITServDf
```
